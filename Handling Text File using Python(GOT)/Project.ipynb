# Extract Names of Each Characters

**According to the format of the text file, name of each character is present at the begining of individual line. This format makes it easy to extract name of the character.**

**In order to achive this: following steps were adopted:**

**1. Read the text line by line (except the blank lines) and store it in the form of a list.**
**2. Extract the unique name of from the individual elements and store them in a seperate list.**

```python
# Read the text file line by line
with open('file.txt', 'r') as file:
    lines = file.readlines()
```

```python
lines
```

**The new line characters are removed and elements are stored in a new list.**

```python
# Remove new line characters
lines_cleaned =[]
for line in lines:
    if line == '\n':
        continue
    else:
        lines_cleaned.append(line[:-1])
```

```python
lines_cleaned
```

**Now, extract the names of the character. From the format of the files, it can be seen that the characters forms the first word of each line.**

**With the exception 'SEPT MORDANE' and 'WAYMA ROYCE' whose full names are used and hence form the fist two character.**

**Also ":" needs to be removed from the ending of the names.**

```python
full_name_ex = ['SEPT','WAYMA']

names = []

for line in lines_cleaned:

    name = line.split()[0][:-1]
    
    if name in full_name_ex:
        first_name = line.split()[0]
        last_name = line.split()[1][:-1]
        name = first_name+ ' ' + last_name 
    
    if name not in names:
        names.append(name)
```

```python
names
```

```python
len(names)
```

**Thus there are 17 unique characters in the text provided.**

# Unique Words By Each Characters

**In order to find theunique words by each character and store them in seperate text file (one word in one line) following steps were adopted:**

   **1. Create a dictionary with names of each character as keys and respective lines as values.**
  
  **2. Create another dictionary that conatins names of the characters as keys and unique words in by the characters as values.**
  
  **3. Write each unique word in a seperate text file inside a directory in present directory.**

```python
# Create a dictionary of names as key and empty lists as values
char_words = {key: [] for key in names}
char_words
```

```python
# Associate each line by the character to respective key in the directory
for name in names:
    
    for line in lines_cleaned:

        
        if name == line.split()[0][:-1]:
            char_words[name].append(line.split()[1:])
    
        elif name.split()[0] == line.split()[0]:
            char_words[name].append(line.split()[2:])
        
```

```python
char_words
```

```python
# Create a dictionary of names as key and empty lists as values
char_unique_words = {key: [] for key in names}
char_unique_words
```

```python
# Add only unique words to the values of the dictionary
for name, lines in char_words.items():
    
    for line in lines:
        
        for word in line:
            
            if word not in list(char_unique_words[name]):
                char_unique_words[name].append(word)
```

```python
# Add only unique words to the values of the dictionary
for name, lines in char_words.items():
    
    for line in lines:
        
        for word in line:
            
            dummy_list = list(char_unique_words[name])
            lowercase_list = [x.lower() for x in dummy_list]
            
            if word.lower() not in lowercase_list:
                char_unique_words[name].append(word)
```

```python
char_unique_words
```

**The above dictionary contains names of the characters as keys and their unique words as values.**

**Now creation of sperate text file for each character and their respective unique words were added in seperate line.**

```python
for name in names:
    
    filename = name +'.txt'
    directory = 'words'
    
    if not os.path.exists(directory):
        os.makedirs(directory)

    with open(os.path.join(directory, filename),'w') as f:
        f.writelines('\n'.join(list(char_unique_words[name])))
```
